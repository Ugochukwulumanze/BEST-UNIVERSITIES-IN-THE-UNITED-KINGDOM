## Student Enrollment by Minimum IELTS and Campus Location
Under this category, we will be looking at Student enrollment according to Universities' minimum IELTS, as well as according the location of these Universites(Urban, Suburban and Rural).

```python
bestuniuk_df=bestuniuk_df.sort_values(by=['UK_rank'],ascending=True)
bestuniuk_df[['University_name','Student_enrollment']]
```

```python
stu_enroll=bestuniuk_df['Student_enrollment'].value_counts()
stu_enroll
```

```python
sns.countplot(y='Student_enrollment', data=bestuniuk_df)
plt.show()
```

Imperial College London and Loughborough University alongside 27 other Universities have in record student enrolment of about 15,000 to 20,000. This is followed by University of East London and University of Bedfordshire with 26 other Universities having in record student enrolment of about 10,000 to 15,000.  

```python
bestuniuk_df=bestuniuk_df.sort_values(by=['Minimum_IELTS_score'],ascending=False)
bestuniuk_df
```

```python
bestuniuk_df[['Student_enrollment_lowerboundary','Student_enrollment_higherboundary']]=bestuniuk_df.Student_enrollment.str.split('-',expand=True)
bestuniuk_df
```

```python
bestuniuk_df['Student_enrollment_lowerboundary']=bestuniuk_df['Student_enrollment_lowerboundary'].str.replace(',','').astype(int)
bestuniuk_df
```

```python
bestuniuk_df['Student_enrollment_higherboundary']=bestuniuk_df['Student_enrollment_higherboundary'].str.replace(',','').astype(int)
bestuniuk_df
```

```python
bestuniuk_df['Stu_enroll_ave']=bestuniuk_df[['Student_enrollment_lowerboundary','Student_enrollment_higherboundary']].mean(axis=1)
bestuniuk_df
```

```python
bestuniuk_df.info()
```

```python
bestuniuk_df['Stu_enroll_ave']=bestuniuk_df['Stu_enroll_ave'].apply(np.int64)
display(bestuniuk_df.dtypes)
```

```python
bestuniuk_df
```

### Student enrollment according to Universities Minimum IELTS score. 

```python
stu_enrol_IELTS=bestuniuk_df[['University_name','Minimum_IELTS_score','Stu_enroll_ave']]
stu_enrol_IELTS
```

```python
stu_enrol_IELTSscore=bestuniuk_df.groupby('Minimum_IELTS_score')['Stu_enroll_ave'].sum()
stu_enrol_IELTSscore
```

```python
stu_enrol_IELTSscore
plt.figure(figsize=(20,10))
plt.title('Student enrollment according to Universities Minimum IELTS score')
plt.pie(stu_enrol_IELTSscore,labels=stu_enrol_IELTSscore.index,autopct='%1.f%%',startangle=180)
```

The Chart above shows that almost 50% of Students enrolled in Unversities with a minimum IELTS Score of 4.5. 42% of the Students that enrolled in Universities in the UK, enrolled in Universities that have a minimum IELTS score of 5.5. Universities with a minimum IELTS score of 6.5 have about 6% of Students enrolled in them.
